```python
import os
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from transformers import BlipProcessor, BlipModel
from datasets import load_dataset, concatenate_datasets
from tqdm.auto import tqdm
from collections import Counter
from sklearn.metrics import accuracy_score, f1_score

# Á¶ÅÊ≠¢ Tokenizers Âπ∂Ë°åË≠¶Âëä
os.environ["TOKENIZERS_PARALLELISM"] = "false"

# ================= ÈÖçÁΩÆÂå∫Âüü (‰ºòÂåñÁâà) =================
BATCH_SIZE = 8
EPOCHS = 20           # Â¢ûÂä†ËΩÆÊï∞ÔºåÂõ†‰∏∫Êàë‰ª¨Èôç‰Ωé‰∫ÜÂ≠¶‰π†ÁéáÔºåÈúÄË¶ÅÊõ¥Â§öÊó∂Èó¥Êî∂Êïõ
LEARNING_RATE = 1e-5  # Èôç‰ΩéÂ≠¶‰π†ÁéáÔºåÈò≤Ê≠¢ Loss ÈúáËç°
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
TOP_K = 1000          # Â¢ûÂä†Ë¶ÜÁõñËåÉÂõ¥ÔºåËÆ©Ê®°ÂûãËÉΩÂ≠¶Âà∞Êõ¥Â§öËØç

SAVE_DIR = "/kaggle/working/blip_finetune"
BEST_MODEL_PATH = f"{SAVE_DIR}/best_model.pt"
os.makedirs(SAVE_DIR, exist_ok=True)

print(f"Ê≠£Âú®‰ΩøÁî®ËÆæÂ§á: {DEVICE}")
# ===================================================

# 1. ÁªòÂõæÂáΩÊï∞ÔºöËÆ≠ÁªÉÊõ≤Á∫ø
def plot_metrics(history):
    sns.set_theme(style="whitegrid")
    epochs = range(1, len(history['train_loss']) + 1)
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10), dpi=100)
    
    # Loss Êõ≤Á∫ø
    ax1.plot(epochs, history['train_loss'], 'o-', color='#6a0dad', label='Train Loss')
    ax1.plot(epochs, history['val_loss'], 'o--', color='orange', label='Val Loss')
    ax1.set_title('Loss Curve')
    ax1.legend()
    
    # ÂáÜÁ°ÆÁéáÊõ≤Á∫ø
    ax2.plot(epochs, history['val_acc'], 's-', color='#008080', label='Total Accuracy')
    ax2.plot(epochs, history['val_closed_acc'], '^-', color='#2e8b57', label='Closed (Yes/No) Acc')
    ax2.plot(epochs, history['val_open_acc'], 'x-', color='#cd5c5c', label='Open Acc')
    
    # Ê†áÊ≥®ÊúÄ‰Ω≥ÁÇπ
    max_acc = max(history['val_acc'])
    best_epoch = history['val_acc'].index(max_acc) + 1
    ax2.annotate(f'Best: {max_acc:.1%}', xy=(best_epoch, max_acc), xytext=(best_epoch, max_acc+0.1),
                 arrowprops=dict(facecolor='black', shrink=0.05))
    
    ax2.set_title('Validation Accuracy Breakdown')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

# 2. ÈöèÊú∫Ê†∑Êú¨ÂèØËßÜÂåñÂáΩÊï∞
def visualize_predictions(model, dataset, processor, device, num_samples=4):
    print(f"\n{'='*20} ÈöèÊú∫Ê†∑Êú¨È¢ÑÊµãÂ±ïÁ§∫ {'='*20}")
    model.eval()
    indices = random.sample(range(len(dataset)), num_samples)
    
    # ÂàõÂª∫ÁîªÂ∏É
    fig, axes = plt.subplots(1, num_samples, figsize=(15, 5))
    if num_samples == 1: axes = [axes]
    
    for i, idx in enumerate(indices):
        item = dataset.dataset[idx] # Ëé∑ÂèñÂéüÂßãÊï∞ÊçÆ
        image = item['image']
        question = item['question']
        ground_truth = str(item['answer']).lower()
        
        # È¢ÑÂ§ÑÁêÜÂπ∂È¢ÑÊµã
        encoding = processor(images=image, text=question, padding="max_length", truncation=True, return_tensors="pt")
        pixel_values = encoding.pixel_values.to(device)
        input_ids = encoding.input_ids.to(device)
        attention_mask = encoding.attention_mask.to(device)
        
        with torch.no_grad():
            _, logits = model(pixel_values, input_ids, attention_mask)
            pred_id = torch.argmax(logits, dim=1).item()
            pred_answer = dataset.id2answer.get(pred_id, "unknown")
            
        # ÁªòÂõæ
        ax = axes[i]
        ax.imshow(image)
        ax.axis('off')
        
        # È¢úËâ≤Âà§Êñ≠ÔºöÂØπÁªøÔºåÈîôÁ∫¢
        color = 'green' if pred_answer == ground_truth else 'red'
        
        title_text = f"Q: {question}\nTrue: {ground_truth}\nPred: {pred_answer}"
        ax.set_title(title_text, fontsize=9, color=color, fontweight='bold')
    
    plt.tight_layout()
    plt.show()

# 3. Êï∞ÊçÆÂáÜÂ§á
print(">>> Ê≠£Âú®Âä†ËΩΩÂπ∂Â§ÑÁêÜÊï∞ÊçÆ...")
raw_dataset = load_dataset("flaviagiammarino/vqa-rad")
full_dataset = concatenate_datasets([raw_dataset['train'], raw_dataset['test']])

# ‰∏•Ê†ºÂàáÂàÜÔºö70% Train, 15% Val, 15% Test

split_1 = full_dataset.train_test_split(test_size=0.3, seed=42)
train_set = split_1['train']
temp_set = split_1['test']


split_2 = temp_set.train_test_split(test_size=0.5, seed=42)
val_set = split_2['train']
test_set = split_2['test']

# ÊûÑÂª∫Á≠îÊ°àÂ∫ì (‰ªÖÂü∫‰∫éËÆ≠ÁªÉÈõÜ)
processor = BlipProcessor.from_pretrained("Salesforce/blip-vqa-base")
all_train_answers = [str(item['answer']).lower() for item in train_set]
answer_counts = Counter(all_train_answers)
top_answers = [ans for ans, count in answer_counts.most_common(TOP_K)]
answer2id = {ans: i for i, ans in enumerate(top_answers)}
id2answer = {i: ans for i, ans in enumerate(top_answers)}
NUM_LABELS = len(top_answers)
print(f"Á≠îÊ°àÁ±ªÂà´Êï∞: {NUM_LABELS}")

class VQADataset(Dataset):
    def __init__(self, dataset, processor, answer2id, id2answer):
        self.dataset = dataset
        self.processor = processor
        self.answer2id = answer2id
        self.id2answer = id2answer

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        item = self.dataset[idx]
        image = item['image']
        question = item['question']
        answer = str(item['answer']).lower()
        encoding = self.processor(images=image, text=question, padding="max_length", truncation=True, return_tensors="pt")
        label = self.answer2id.get(answer, -1) 
        return {
            "input_ids": encoding.input_ids.squeeze(),
            "pixel_values": encoding.pixel_values.squeeze(),
            "attention_mask": encoding.attention_mask.squeeze(),
            "labels": torch.tensor(label, dtype=torch.long),
            "original_answer": answer
        }

# ÂÆû‰æãÂåñÊï∞ÊçÆÈõÜ
train_ds = VQADataset(train_set, processor, answer2id, id2answer)
val_ds = VQADataset(val_set, processor, answer2id, id2answer)
test_ds = VQADataset(test_set, processor, answer2id, id2answer)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False)

# 4. Ê®°ÂûãÂÆö‰πâ
class BLIP_VQA_Classifier(nn.Module):
    def __init__(self, num_answers):
        super().__init__()
        self.blip = BlipModel.from_pretrained("Salesforce/blip-vqa-base")
        # ÂÜªÁªìËßÜËßâÂ±Ç
        for param in self.blip.vision_model.parameters():
            param.requires_grad = False
        
        # ÂàÜÁ±ªÂ§¥
        self.classifier = nn.Sequential(
            nn.Linear(self.blip.config.text_config.hidden_size, 512),
            nn.ReLU(),
            nn.Dropout(0.3), # Á®çÂæÆÂ¢ûÂä† Dropout
            nn.Linear(512, num_answers)
        )

    def forward(self, pixel_values, input_ids, attention_mask, labels=None):
        outputs = self.blip(
            pixel_values=pixel_values,
            input_ids=input_ids,
            attention_mask=attention_mask,
            return_dict=True
        )
        pooled_output = outputs.text_model_output.last_hidden_state[:, 0, :]
        logits = self.classifier(pooled_output)
        loss = None
        if labels is not None:
            loss = nn.CrossEntropyLoss()(logits, labels)
        return loss, logits

model = BLIP_VQA_Classifier(num_answers=NUM_LABELS).to(DEVICE)
optimizer = torch.optim.AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=LEARNING_RATE)
# Êñ∞Â¢ûÔºöÂ≠¶‰π†ÁéáË∞ÉÂ∫¶Âô® (Â¶ÇÊûú Val Loss ‰∏çÈôçÔºåÂ∞±Èôç‰Ωé LR)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2, verbose=True)

# ================= 5. ËÆ≠ÁªÉ‰∏ªÂæ™ÁéØ =================
print("\n>>> ÂºÄÂßãËÆ≠ÁªÉ (ÂåÖÂê´ Open/Closed ËØ¶ÁªÜËØÑ‰º∞)...")
history = {'train_loss': [], 'val_loss': [], 'val_acc': [], 'val_closed_acc': [], 'val_open_acc': []}
best_val_f1 = 0.0

for epoch in range(EPOCHS):
    # --- Training ---
    model.train()
    train_loss = 0
    for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}"):
        input_ids = batch["input_ids"].to(DEVICE)
        pixel_values = batch["pixel_values"].to(DEVICE)
        attention_mask = batch["attention_mask"].to(DEVICE)
        labels = batch["labels"].to(DEVICE)
        
        valid = labels != -1
        if not valid.any(): continue
        
        optimizer.zero_grad()
        loss, _ = model(pixel_values[valid], input_ids[valid], attention_mask[valid], labels[valid])
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
    
    avg_train_loss = train_loss / len(train_loader)
    
    # --- Validation (ËØ¶ÁªÜÁâà) ---
    model.eval()
    val_loss = 0
    all_preds, all_labels = [], []
    closed_preds, closed_labels = [], []
    open_preds, open_labels = [], []
    
    with torch.no_grad():
        for batch in val_loader:
            input_ids = batch["input_ids"].to(DEVICE)
            pixel_values = batch["pixel_values"].to(DEVICE)
            attention_mask = batch["attention_mask"].to(DEVICE)
            labels = batch["labels"].to(DEVICE)
            
            # ËÆ°ÁÆó Val Loss
            valid = labels != -1
            if valid.any():
                batch_loss, logits = model(pixel_values[valid], input_ids[valid], attention_mask[valid], labels[valid])
                val_loss += batch_loss.item()
            else:
                _, logits = model(pixel_values, input_ids, attention_mask)
                
            preds = torch.argmax(logits, dim=1)
            
            # ÁªüËÆ°ÂàÜÁ±ªÊåáÊ†á
            for p, l in zip(preds.cpu().numpy(), labels.cpu().numpy()):
                if l == -1: continue
                all_preds.append(p)
                all_labels.append(l)
                
                ans_str = id2answer[l]
                if ans_str in ['yes', 'no']:
                    closed_preds.append(p)
                    closed_labels.append(l)
                else:
                    open_preds.append(p)
                    open_labels.append(l)

    # ËÆ°ÁÆóÂêÑÈ°πÊåáÊ†á
    avg_val_loss = val_loss / len(val_loader) if len(val_loader) > 0 else 0
    val_acc = accuracy_score(all_labels, all_preds)
    val_f1 = f1_score(all_labels, all_preds, average='weighted')
    
    acc_closed = accuracy_score(closed_labels, closed_preds) if closed_labels else 0
    acc_open = accuracy_score(open_labels, open_preds) if open_labels else 0
    
    # Êõ¥Êñ∞ History
    history['train_loss'].append(avg_train_loss)
    history['val_loss'].append(avg_val_loss)
    history['val_acc'].append(val_acc)
    history['val_closed_acc'].append(acc_closed)
    history['val_open_acc'].append(acc_open)
    
    # ÊâìÂç∞ËØ¶ÁªÜÊä•Âëä
    print(f" -> Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")
    print(f" -> Overall Acc: {val_acc:.2%} (F1: {val_f1:.4f})")
    print(f" -> [Closed] Yes/No Acc: {acc_closed:.2%}")
    print(f" -> [Open] Other Acc:    {acc_open:.2%}")
    
    # Ë∞ÉÊï¥Â≠¶‰π†Áéá
    scheduler.step(avg_val_loss)

    # ‰øùÂ≠òÊúÄ‰Ω≥Ê®°Âûã
    if val_f1 > best_val_f1:
        best_val_f1 = val_f1
        torch.save(model.state_dict(), BEST_MODEL_PATH)
        print(" -> üéâ Model Saved!")
    print("-" * 40)

# ================= 6. ÁªìÊûúÂ±ïÁ§∫ =================
print("\n>>> ËÆ≠ÁªÉÁªìÊùüÔºåÊ≠£Âú®ÁîüÊàêÂõæË°®...")
plot_metrics(history)

# ÊúÄÁªàÊµãËØïÈõÜËØÑ‰º∞
print("\n>>> Ê≠£Âú®ËøõË°åÊúÄÁªàÊµãËØïÈõÜËØÑ‰º∞...")
model.load_state_dict(torch.load(BEST_MODEL_PATH)) # Âä†ËΩΩÊúÄÂ•ΩÁöÑÈÇ£‰∏™
model.eval()

# ÂèØËßÜÂåñ 4 ‰∏™ÈöèÊú∫Ê†∑Êú¨
visualize_predictions(model, test_ds, processor, DEVICE, num_samples=4)
```

```python
import math

def visualize_predictions(model, dataset, processor, device, num_samples=6, cols=3):
    print(f"\n{'='*20} ÈöèÊú∫Ê†∑Êú¨È¢ÑÊµãÂ±ïÁ§∫ (ÊØèË°å {cols} Âº†) {'='*20}")
    model.eval()
    
    # ÈöèÊú∫ÊäΩÂèñÊ†∑Êú¨
    indices = random.sample(range(len(dataset)), num_samples)
    
    # ËÆ°ÁÆóÈúÄË¶ÅÁöÑË°åÊï∞
    rows = math.ceil(num_samples / cols)
    
    # ÂàõÂª∫ÁîªÂ∏ÉÔºöÈ´òÂ∫¶ÈöèË°åÊï∞Ëá™Âä®Â¢ûÂä†ÔºåÈò≤Ê≠¢Êå§Âéã
    fig, axes = plt.subplots(rows, cols, figsize=(18, 6 * rows))
    
    # Êää axes Â±ïÂπ≥Êàê‰∏ÄÁª¥ÂàóË°®ÔºåÊñπ‰æøÈÅçÂéÜ
    if num_samples > 1:
        axes = axes.flatten()
    else:
        axes = [axes]
    
    for i, idx in enumerate(indices):
        item = dataset.dataset[idx] # Ëé∑ÂèñÂéüÂßãÊï∞ÊçÆ
        image = item['image']
        question = item['question']
        ground_truth = str(item['answer']).lower()
        
        # È¢ÑÂ§ÑÁêÜÂπ∂È¢ÑÊµã
        encoding = processor(images=image, text=question, padding="max_length", truncation=True, return_tensors="pt")
        pixel_values = encoding.pixel_values.to(device)
        input_ids = encoding.input_ids.to(device)
        attention_mask = encoding.attention_mask.to(device)
        
        with torch.no_grad():
            _, logits = model(pixel_values, input_ids, attention_mask)
            pred_id = torch.argmax(logits, dim=1).item()
            pred_answer = dataset.id2answer.get(pred_id, "unknown")
            
        # ÁªòÂõæ
        ax = axes[i]
        ax.imshow(image)
        ax.axis('off')
        
        # ÁªìÊûúÂà§Êñ≠È¢úËâ≤
        is_correct = pred_answer == ground_truth
        color = 'green' if is_correct else 'red'
        icon = "‚úÖ" if is_correct else "‚ùå"
        
        # ËÆæÁΩÆÊ†áÈ¢ò (Â¢ûÂä†Êç¢Ë°åÔºåÈò≤Ê≠¢ÊñáÂ≠óÈáçÂè†)
        title_text = f"{icon} Q: {question}\nTrue: {ground_truth}\nPred: {pred_answer}"
        ax.set_title(title_text, fontsize=12, color=color, fontweight='bold', pad=10)
    
    # Â¶ÇÊûúÊ†∑Êú¨Êï∞Â°´‰∏çÊª°ÊúÄÂêé‰∏ÄË°åÔºåÊääÂ§ö‰ΩôÁöÑÁ©∫ÂõæÂÖ≥Êéâ
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()

# --- ËøêË°åÂëΩ‰ª§ ---
# Áé∞Âú®Â±ïÁ§∫ 6 Âº†ÂõæÔºåÊØèÊéí 3 Âº†
visualize_predictions(model, test_ds, processor, DEVICE, num_samples=6, cols=3)
```
